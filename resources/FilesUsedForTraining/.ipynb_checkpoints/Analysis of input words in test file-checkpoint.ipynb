# Chinese word segmentation
## Analysis of input words in test file


```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
datasets_names = {'as','cityu','msr','pku'}
sents_orig = {}
for name in datasets_names:
    with open('../dataset/icwb2-data/testing/%s_test.utf8' % name, 'r', encoding='utf-8') as file:
        sents_orig[name] = file.readlines()
    sents_orig[name] = [x.split('\n')[0] for x in sents_orig[name]] # Remove the new line character
```

```python
exceeded_orig=[]
total_sentences = []
for name in datasets_names:
    for s in sents_orig[name]:
        total_sentences.append(s)
        if len(s) > 50:
            exceeded.append(s)
```

```python
cutChar = re.compile(u'[。，、？！\.\?,!；：”“]')
j = 0
sents_cut ={}
cuts_in_sents = {} # Saves the position of the cuts, to rebuild later
is_cut=False
for name in datasets_names:
    sents_cut[name] = []
    cuts_in_sents[name]=[]
    for sent in sents_orig[name]:
        cuts = []
        j = 0
        for i in cutChar.finditer(sent):
            cuts.append(i.start()) # Appends the position of the cut
            is_cut=True
            sents_cut[name].append(sent[j:i.start()])
            j = i.end()
        if is_cut:
                if j <= len(sent)-1:
                    sents_cut[name].append(sent[j:len(sent)])
                    cuts.append(j)
        if not is_cut:
            sents_cut[name].append(sent)
        is_cut = False
        cuts_in_sents[name].append(cuts)
    
```

```python
sents_cut[name]
```

```python
lens=[]
for name in datasets_names:
    lens.append([len(x) for x in sents_cut[name]])
plt.boxplot(lens)
plt.gcf()
plt.savefig("../figures/cut_sentences.svg")
```

```python
lens=[]
for name in datasets_names:
    lens.append([len(x) for x in sents_orig[name]])
plt.boxplot(lens)
plt.gcf()
plt.savefig("../figures/test_sents.svg")
```

```python
exceeded=[]
total_sentences = []
for name in datasets_names:
    for s in sents_cut[name]:
        total_sentences.append(s)
        if len(s) > 50:
            exceeded.append(s)
```

```python
print("Number of sentences with length higher than 50:" ,len(exceeded))
print("Total number of sentences:" ,len(total_sentences))
print("Porcentage relationship: ", len(exceeded)/len(total_sentences)*100,'%' )
```
